```python
import numpy as np
import sys
np.set_printoptions(precision=4)
np.set_printoptions(threshold=30)
np.set_printoptions(linewidth=300)
np.set_printoptions(suppress=True)
np.set_printoptions(threshold=sys.maxsize)
```

```python
import matplotlib.pyplot as plt
X = np.array([[1,1],
             [2,1],
             [1.5, 0],
             [3, 2]])

y = np.array([[1],
             [1],
             [0],
             [2]])
phi = np.array([[1, 1, 1],
               [4, 2, 1],
               [2.25, 1.5, 1],
               [9, 3, 1]])

w = np.linalg.inv(phi.T.dot(phi)).dot(phi.T).dot(y)
print(w)
```

```python
#Made with chat gpt

x_values = X[:, 0]
y_values = y

# Plot the data points
plt.scatter(x_values, y_values, color='blue', label='Data Points')

# Calculate the vertex of the parabola
vertex_x = -w[1] / (2 * w[0])

# Generate x values for the line plot, centered around the vertex
x_line = np.linspace(vertex_x - 2, vertex_x + 2, 100)

# Generate y values for the line plot using the weights w
y_line = w[0] * x_line**2 + w[1] * x_line + w[2]

# Plot the line
plt.plot(x_line, y_line, color='red', label='Fitted Line')

# Adding labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data Points and Fitted Line Centered at Vertex')
plt.legend()

# Show the plot
plt.show()
```
