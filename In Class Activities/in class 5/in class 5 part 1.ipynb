```python
import numpy as np
from numpy import ones
from numpy import array
import matplotlib.pyplot as plt
from numpy import mean
```

Given the Objective <br>
$$ \min_{\alpha, \beta} \; 5\alpha^2 + 6 \alpha \beta - 16 \alpha + 3 \beta^2 - 12 \beta + 14$$

The derivative is <br>
$$ \begin{cases} \frac{df}{d\alpha} = 10\alpha + 6\beta - 16 \\  \frac{df}{d\alpha} = 6\alpha + 6\beta - 12  \end{cases} = \begin{bmatrix} 10 & 6  \\ 6 & 6 \end{bmatrix} \begin{bmatrix} \alpha \\ \beta \end{bmatrix} - \begin{bmatrix} 16 \\ 12 \end{bmatrix} = A w - c$$<br>
Let 

```python
A = array([[10,6],[6,6]])
c = array([[16],[12]])
w = array([[2],[2]]) # solution is 1,1
η = 0.1
```

```python
def f(w):
	α = w[0]
	β = w[1]
#	
	return 5*α*α + 6*α*β - 16*α + 3*β*β - 12*β + 14
```

```python
f_value_list = []
for i in range(20):
	fᑊ = A.dot(w) - c
	w = w - η*fᑊ				# gradient descent update w
	f_value_list.append(f(w))
```

```python
print('Best w = \n', w)
plt.plot(f_value_list)
plt.title('MSE over GD')
plt.xlabel('steps')
plt.ylabel('MSE')
plt.show()
```

## Minimize the following 2 problems:

```python
x1 = np.array([[0], [1]])
x2 = np.array([[1], [1]])
x = np.vstack((x1.T, x2.T))
y = np.array([[0], [2]])
n=2
w = array([[2],[2]])
lr = 0.1
```

```python
def f(w):
    num = 0
    for xi, yi in zip(x,y):
        xi = np.reshape(xi, (2,1))
        num += (w.T.dot(xi) - yi)**2
    return (num / n).item()
```

```python
def f_dx(w):
    num = 0
    for xi, yi in zip(x,y):
        xi = np.reshape(xi, (2,1))
        num += (w.T.dot(xi) - yi)*(xi)
    return (2/n) * num
```

```python
f_value_list = []
for i in range(20):
    w = w - lr * f_dx(w)
    f_value_list.append(f(w))
```

```python
print('Best w = \n', w)
```

```python
plt.plot(f_value_list)
plt.title('MSE over GD')
plt.xlabel('steps')
plt.ylabel('MSE')
plt.show()
```

```python
A = np.array([[0, 1], [1, 1]])
y = np.array([[0], [2]])
w = array([[2],[2]])
lr = 0.1
```

```python
def f(w):
    temp = np.dot(A,w) - y
    return np.dot(np.transpose(temp), temp).item()
```

```python
f_value_list = []
def f_dx(w):
    return 2*A.T.dot(A.dot(w) - y) 
```

```python
for i in range(100):
    w = w - lr * f_dx(w)
    f_value_list.append(f(w))
```

```python
print('Best w = \n', w)
```

```python
plt.plot(f_value_list)
plt.title('MSE over GD')
plt.xlabel('steps')
plt.ylabel('MSE')
plt.show()
```
