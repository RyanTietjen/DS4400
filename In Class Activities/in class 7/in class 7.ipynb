```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

```python
lunch_wait_time = genfromtxt('lunch_wait_time.csv', delimiter = ',')
SAT = genfromtxt('SAT.csv', delimiter = ',')
student_age = genfromtxt('student_age.csv', delimiter = ',')
```

```python
plt.hist(lunch_wait_time, 20)
plt.title("Lunch wait time")
print("Average wait time: ", np.mean(lunch_wait_time))
wait_time_count = 0
for i in lunch_wait_time:
    if i > 3:
        wait_time_count += 1
print("Probability of a wait time > 3 minutes: ", wait_time_count/len(lunch_wait_time))
plt.show()
```

```python
plt.hist(SAT, 20)
print("Average SAT: ", np.mean(SAT))
plt.title("SAT")
SAT_count = 0
for i in SAT:
    if i > 1300:
        SAT_count += 1
print("Probability of a SAT > 1300: ", SAT_count/len(SAT))
plt.show()
```

```python
plt.hist(student_age, 20)
plt.title("Age of random students")
print("Average age: ", np.mean(student_age))
age_count = 0
for i in student_age:
    if i < 20:
        age_count += 1
print("Probability of someone being younger than 20: ", age_count/len(student_age))
plt.show()
```

```python
from scipy.stats import norm
print("To be in the 95th percentile, you would need a score of", norm.ppf(0.95, 1000, 200))
```

```python
print("Percentage of males taller than 67: ", 1 - norm.cdf(67, 70, 3))
print("Percentage of females shorter than 67: ", norm.cdf(67, 64.5, 2.5))
print("90th percentile of female height: ", norm.ppf(.90, 64.5, 2.5))
```

```python
x_males = np.linspace(55, 84, 100)
x_females = np.linspace(50, 79, 100)

y_males = norm.pdf(x_males, 70, 3)
y_females = norm.pdf(x_females, 64.5, 2.5)

x_fill = np.linspace(35, 67, 100)
aoc_shaded = norm.pdf(x_fill, 64.5, 2.5)
```

```python
plt.plot(x_males, y_males)
plt.plot(x_females, y_females)
plt.fill_between(x_fill, aoc_shaded,  alpha= 0.5, color="grey")
```
